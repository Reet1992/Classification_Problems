```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('listings.csv')
```

```python
df.head()
```

```python
df['affordability'] = (df['price'] * df['minimum_nights'])/df['availability_365']
```

```python
df.dropna()
```

```python
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
room=le.fit_transform(df['room_type'])
df['room_type'] = room
```

```python
df
```

```python
ng=le.fit_transform(df['neighbourhood_group'])
df['neighbourhood_group'] = ng
```

```python
df
```

```python

feature_cols = ['neighbourhood_group','price']
x = df[feature_cols]
y= df['room_type']
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/12,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
